Day 40 – List Methods and Slicing


```python
# Create a 2D list where each row holds a contact’s name, phone number, and email address.
contact = [
    ["Abi",1234567890,"abi@gmail.com"],
    ["Banu",9987654321,"banu@gmail.com"],
    ["Vishnu",3456789012,"vishnu@gmail.com"],
    ["Subesh",4567890123,"subesh@gmail.com"]
]
print(contact)
```

```python
# Allow the user to add a new contact.
name = input("Enter the name: ")
phone = int(input("Enter the phone number: "))
email = input("Enter the email address: ")
contact.append([name,phone,email])
print(contact)
```

```python
# Display all contacts sorted alphabetically by name.
contact.sort()
print(contact)
```

```python
# Allow the user to search for contacts by name. The search should support partial matches
name1 = input("Enter name: ").lower()
for i in contact:
  if name1 in i[0].lower():
      print(i)
```

```python
# After sorting, display only the first 3 contacts in the list.
contact.sort()
print(contact[:3])
```
